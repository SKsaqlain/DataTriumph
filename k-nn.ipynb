IMPORTING THE REQUIRED MODULES


```python
import pandas as pd
import numpy as np
```

IMPORTING THE DATASETS

```python
dataset=pd.read_csv('dataset/processed_1.csv')
dataset.head()
```

```python
#IMPORITNG OLD UNPROCESSED DATASET TO GET THE UNIQUE ITEMS THAT ARE THERE IN THE DATASET.
old_dataset=pd.read_csv(r'dataset/BreadBasket_DMS.csv')
#old_dataset=dataset.mask(old_dataset.eq('NONE')).dropna()
print(old_dataset.head())
unique_set_items=set(old_dataset.Item)
print(len(unique_set_items))
```

```python
items_index=list(unique_set_items)
print(items_index)
```

CONVERTING THE TRANSACTIONS TO VECTORS

```python
x=list()
y=list()
#print(len(dataset.Items))
for i in range(len(dataset.Items)):
    itm=dataset.Items[i].split("|")
    #print(itm)
    if(len(itm)>1):
        for k in range(0,len(itm)-1):
            x_vec=[0]*95
            for j in range(0,k+1):
                x_vec[items_index.index(itm[j])]=1
            print(itm[0:k+1],'->',itm[k+1])
            x.append(x_vec)
            y.append(items_index.index(itm[k+1]))
```

IMPORTING THE REQUIRES LIBRARIES

```python
from sklearn import  model_selection
#splitting the datset into training and testing set.
X_train,X_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#initialising knn classifier
knn_clf=KNeighborsClassifier()
knn_clf.fit(X_train,y_train)
```

```python
#fitting the model and calculating the accuracy
y_pred=knn_clf.predict(X_test)
count=0
for i in range(len(y_pred)):
    if(y_pred[i]==y_test[i]):
        count+=1
print(count/len(y_pred))
```
