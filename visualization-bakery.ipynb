    <li>Most Bought Itemsets</li>

**in the bakery dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly as py
import warnings
warnings.filterwarnings("ignore")
data = pd.read_csv(r"BreadBasket_DMS.csv").drop_duplicates()
processed = pd.read_csv(r"processed_1.csv")
data=data.mask(data.eq('NONE')).dropna()

#finding unique items

# unique_set_items=set(data.Item)
# print(unique_set_items)
# print(len(unique_set_items))

```

```python
#most bought 
counts=data['Item'].value_counts()
top_ten=counts[:10]

print(top_ten)

#top_ten.plot()
top_ten.plot(kind='bar',title="Top Ten Most Bought Items",rot=45,figsize=(10,10))
plt.show()
```

```python
#finding unique itemsets

unique_set_items=set(processed.Items)
#print(unique_set_items)

#most common itemsets
itemset_counts=processed['Items'].value_counts()
itemset_top_ten=itemset_counts[:10]

print(itemset_top_ten)

#top_ten.plot()
itemset_top_ten.plot(kind='bar',title="Top Ten Itemsets",rot=45,figsize=(10,10))
plt.show()
```

```python
#finding unique itemsets

unique_set_items=set(processed.Items)
#print(unique_set_items)

#most common itemsets
itemsets=pd.DataFrame(columns=['Item'])
item=processed['Items']
i=0;
for x in item:
    if(len(x.split('|'))>1):
        itemsets.loc[i]=x
        i+=1
itemset_counts=itemsets['Item'].value_counts()

top=itemset_counts[:10]

#top_ten.plot()
top.plot(kind='bar',title="Top Ten Itemsets",rot=45,figsize=(10,10))
plt.show()
```
